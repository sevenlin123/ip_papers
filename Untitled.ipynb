```python
%matplotlib inline
import os
import json
from matplotlib import pyplot as plt
import plotly.plotly as py
from wordcloud import WordCloud, STOPWORDS
from numpy import *
from PIL import Image
```

```python
fn = 'C302_20170610-1.txt'
f = open(fn).read()
jl = []
jd = {}
fa = 0
dap = {}
for nn, i in enumerate(f.split('\r')):
    i = i.split('\t')
    al, yr, ti, jn = i[0], i[1], i[2], i[3]
    dap[nn]={'al':al.replace('\xca', ' ').strip('"').strip().split(';'), 'year':yr, 'title':ti.replace('67P/Churyumov-Gerasimenko', '67P'), 'jn':jn}
```

```python
jn_count = {}
al_count = {}
title_count = {}
year_count = {}
text = ''
text_67p = ''
title = ''
jn = ''
for i in dap:
    try:
        jn_count[dap[i]['jn']] +=1
    except KeyError:
        jn_count[dap[i]['jn']] = 1
    jn += dap[i]['jn']
    jn += ' '
    try:
        year_count[dap[i]['year']] +=1
    except KeyError:
        year_count[dap[i]['year']] = 1
    title += dap[i]['title'].strip().strip('"')
    title += ' '
    for j in dap[i]['al']:
        if len(j.split(','))> 1:
            j = j.split(',')[0].strip()
            try:
                al_count[j] +=1
            except KeyError:
                al_count[j] = 1
            if j != 'Ip' and  j != 'De' and j!= 'de' and len(j) > 1 and al_count[j] <20:
                if dap[i]['title'].find('67P') == -1:
                    text += j
                    text += ' '
                    text_67p += j
                    text_67p += ' '
                else:
                    text_67p += j
                    text_67p += ' '
                
```

```python
ip_mask = array(Image.open('Ip.png'))
photo_mask = array(Image.open('photo.jpg'))
```

```python
wc = WordCloud(background_color="white", max_words=2000, margin=0,
               random_state=1, width=900, height = 500, mask=ip_mask, colormap='plasma').generate(text)
```

```python
default_colors = wc.to_array()
plt.figure(figsize=(15,15))
plt.axis("off")
plt.imshow(wc.recolor(random_state=4),
           interpolation="bilinear")
plt.savefig('authors.png', dpi=150)
plt.show()
```
